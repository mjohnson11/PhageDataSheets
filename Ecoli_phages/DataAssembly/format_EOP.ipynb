```python
import pandas as pd

eop = pd.read_csv('../../../Moriniere_2026_Metadata/KEIO_EOP.csv')
eop['Genotype'] = eop.strain.apply(lambda x: x.split('-pCA')[0].split('::Y')[0])
eop_non_comp = eop[eop.strain.str.endswith('24N')].rename(columns={'values':'EOP'})
eop_comp = eop[eop.strain.str.endswith(')')].rename(columns={'values':'EOP (complemented)'})

eop_reformatted = eop_non_comp[[i for i in eop_non_comp if i!='strain']].merge(eop_comp[[i for i in eop_comp if i!='strain']], on=['phage', 'Genotype'])[['phage', 'Genotype', 'EOP', 'EOP (complemented)']]
eop_reformatted.to_csv('../data/KEIO_EOP_reformatted.csv', index=False)
eop_reformatted
```
