```python
import pandas as pd
import json
with open('../../../Moriniere_2026_Metadata/EDGE_network.cyjs') as f:
    data = json.load(f)
nodes = data['elements']['nodes']
mat = []
for node in nodes:
    mat.append([node['data']['name'], node['data']['id'], node['position']['x'], node['position']['y']])
    
df = pd.DataFrame(mat, columns=['Phage', 'ID', 'x', 'y'])
df.to_csv('../data/network_nodes.csv', index=False)

mat = []
for edge in data['elements']['edges']:
    mat.append([edge['data']['source'], edge['data']['target']])
    
df = pd.DataFrame(mat, columns=['source', 'target'])
df.to_csv('../data/network_edges.csv', index=False)
```
